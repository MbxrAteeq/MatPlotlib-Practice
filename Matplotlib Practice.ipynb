# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

## 1. Functional Method

```python
# x argue is the data on x-axis and Y argue is the data on y-axis
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
# Multiple plots on same Canvas
# plt.subplot (Number of rows,Number of columns,plot number)
# Number of rows are the rows will show in outout
# Number of columns are columns will show in output
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

## 2. Object Oriented Api Method

```python
fig = plt.figure()
# We created a figure

axes = fig.add_axes([0.1,0.1,0.8,0.8])
# We called the mothod add axis on that figure and passes a list of axes
# Parameteres are ([left of axes,bottom of axes,width,height])

axes.plot(x,y)
# Then we plotted the data on axis

# setting labels on both axes and title
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

### Two sets of figures on one Canvas

```python
#In this multiple figure plotting you manually add the location/axis
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Title 1')
axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')

axes2.plot(y,x,'r')
axes2.set_title('Title 2')
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
```

### SubPlots

```python
#In this multiple figure plotting you manually add the location/axis

fig2,axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
#axes here are actually an array
axes[0].plot(x,y)
axes[0].set_title("First Plot")


axes[1].plot(x,y)
axes[1].set_title("First Plot")


# plt.tight_layout() # For layout issues if we have more columns and rows
```

### Figure Size, Aspect Ratio, DPI

```python
fig3 = plt.figure(figsize=(4,4))
# we will pass tuple to figsize in which (width,height) of figure will bi in inches.

ax = fig3.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
# We also can add figsize to subpots
fig5,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))
#axes here are actually an array
axes[0].plot(x,y)
axes[0].set_title("First Plot")


axes[1].plot(x,y)
axes[1].set_title("First Plot")
```

### How to Save Figure

```python
# You can save in jpeg, png, pdf and etc.
fig5.savefig('MyPicture.png')
```

### Legends

```python
fig0 = plt.figure()

ax = fig0.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X-Squre') #ploting x and x-squre
ax.plot(x,x**3,label='X-Cube') #ploting x and x-cube

ax.legend()

#ax.set_xlabel("X")
#ax.set_ylabel("Y")
#ax.set_title("Title")
```

```python
# Legtend in subplot
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4))

axes[0].plot(x,x**2,label="X-squre")
axes[0].plot(x,x**3,label="X-cube")

axes[1].plot(x,y)

axes[0].legend()
```

```python
fig.savefig('myFig.png')
```

### Setting colors, Line type, Line width

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='Green',lw=4,ls='--',marker='o',markersize=10,markerfacecolor='black',markeredgewidth=3,markeredgecolor='red') #alpha=0.5 you can add for line transpernacy
#lw for linewidth and alpha for transperancy Of line
#ls is for linestyle you can chose '-','--' and etc.

#Marker is the points where our data lies
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',lw=2,ls='--')
```
